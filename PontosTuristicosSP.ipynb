```python
# Ponto central do mapa
center_lat = -14.24007
center_lon = -53.18050
```

```python
consulta = list(pontosSP.find())
```

```python
# Convertendo os dados de consulta em data frame
df_convert = pd.DataFrame(consulta)
```

```python
# imprimindo primeira linha do data frame
df_convert.iloc[0]
```

```python
# Criando um mapa a partir do ponto central com 800x600
locationmap = folium.Map(location=[center_lat , center_lon ], zoom_start=5, width=800, heigth=600 )
```

```python
# Retornando quantidade de pontos
qtd = len(df_convert)
```

```python
qtd
```

```python
for i in range(qtd):
  lng = float(df_convert.iloc[i]['Longitude'])
  lat = float(df_convert.iloc[i]['Latitude'])
  descricao = 'Ponto Turístico: ' + str(df_convert.iloc[i]['Ponto Turístico']) + '<br> Cidade: ' +  str(df_convert.iloc[i]['Cidade']) +'<br> Latitude: ' +  str(df_convert.iloc[i]['Latitude']) + '<br> Longitude: ' +  str(df_convert.iloc[i]['Longitude']) + '<br> : Média Anual de Visitação: ' +  str(df_convert.iloc[i]['Média Anual de Visitação'])
  folium.Marker ( [ lat , lng ],
                popup= descricao,
                 icon=folium.Icon(color='blue')).add_to(locationmap)
```
